```python
#Importing libraries
import pandas as pd
import numpy as np
```

```python
#Loading the dataset
data = pd.read_csv('C:\\Users\\user\\Desktop\\Blackcoffer\\data.csv')

#Setting seed for reproducibility
np.random.seed(0)
```

```python
#Looking at first 10 rows of the dataset
data.head(20)
```

```python
# Calculate the mode of start_year and median of end_year for the entire DataFrame
mode_of_start_year = data['start_year'].mode()[0]
median_of_end_year = data['end_year'].median()
mode_of_end_year = data['end_year'].mode()[0]
print(median_of_end_year)

# Replace the null values in start_year in the 'Energy' sector with the calculated mode
# Replace the null values in end_year in the 'Energy' sector with the calculated median
data.loc[data['start_year'].isnull(), 'start_year'] = mode_of_start_year
data.loc[(data['sector'] == 'Energy') & (data['end_year'].isnull()), 'end_year'] = median_of_end_year
```

```python
#data.head(50)
```

```python
total_cells = data.size
print('total cells:', total_cells)

#Getting the number of missing datapoints per column
missing_values_count = data.isnull().sum()

#Calculating total missing values
total_missing = missing_values_count.size
print('total missing values:',total_missing)

#Percentage of missing values
percent_missing = (total_missing/total_cells) * 100
print('Percentage of missing values:',percent_missing)
```

```python
# Calculate the mean of end_year for 'Aerospace and defence' sector
aero_def_mean_end_year = data[data['sector'] == 'Aerospace & defence']['end_year'].mean()

# Replace the null end_year values in 'Aerospace and defence' sector with the calculated mean
data.loc[(data['sector'] == 'Aerospace & defence') & (data['end_year'].isnull()), 'end_year'] = int(round(aero_def_mean_end_year))
```
